```python
# Import modules
import csv, json
import pandas as pd
```

## Con Ed SIR Inventory

```python
# CSV and JSON paths
coned_all_csvPath = 'data/clean/csv/coned_all_df.csv'
coned_all_jsonPath = 'data/clean/json/coned_all.json'
```

```python
# Read CSV and add to empty dictionary 'data'
data = {}
with open(coned_all_csvPath) as csvFile:
    csvReader = csv.DictReader(csvFile)
    for rows in csvReader:
        developer = rows['developer']
        data[developer] = rows
```

```python
# Create JSON file and write data to
with open(coned_all_jsonPath, 'w') as jsonFile:
    jsonFile.write(json.dumps(data, indent=4))
```

```python
# Check CSV
coned_all_csv = pd.read_csv('data/clean/csv/coned_all_df.csv')
coned_all_csv.head(3)
```

```python
# Check JSON
coned_all_json = pd.read_json('data/clean/json/coned_all.json')
coned_all_json.head()
```
